### ðŸ”¹ Q9. Create a rolling average of 'Stock\_Price' over a window of 3 days.

```python
import pandas as pd
df = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=10),
    'Stock_Price': [100, 102, 101, 105, 110, 108, 107, 109, 111, 112]
})
df.set_index('Date', inplace=True)
```

```python
df
```

```python
df['rolling_avg']= df['Stock_Price'].rolling(window=3,min_periods=1).mean()
```
